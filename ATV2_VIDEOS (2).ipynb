```python
#primeira modificação, não defini um tamanho minimo das mudanças

import cv2
import numpy as np
from IPython.display import display, Image

path = 'C:/Users/carlo/TETI/Videos/'

cap = cv2.VideoCapture(path + 'mont_russa.mp4')

if not cap.isOpened():
    print('Erro na leitura...')
    exit()

# Armazenar a cópia anterior do frame
prev_frame = None

v_contframes = 0
while True:
    ret, frame = cap.read()
    
    if not ret:
        print('Finalizou o vídeo')
        break
        
    # Converter o frame para HSV
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Checar os pixels e sinalizar quando um mudar
    if prev_frame is not None:
        diff = cv2.absdiff(hsv, prev_frame)
        gray = cv2.cvtColor(diff, cv2.COLOR_BGR2GRAY)
        thresh = cv2.threshold(gray, 25, 255, cv2.THRESH_BINARY)[1]
        contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
        
        # Desenhar caixas delimitadoras nos pixels alterados
        for contour in contours:
            x, y, w, h = cv2.boundingRect(contour)
            cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

    # Exibir o frame
    _, img_encoded = cv2.imencode('.jpg', frame)
    display(Image(data=img_encoded))
    
    # Atualizar a cópia anterior do frame
    prev_frame = hsv.copy()
    
    v_contframes += 1
    
cap.release()
print('Total de frames: ', v_contframes)
```
